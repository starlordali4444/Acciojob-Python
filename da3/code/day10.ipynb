    Data Check

    ndim => gives the dimention of the dataframe
    shape=> gives the shape
    size => size of the df
    dtype => data type of the df
    df.info() => 
        will give 
            no of rows
            no of columns
            list of colums/ non null values in each columns/ data type of each column
    df.describe() 
        will give statistical values for numerical columns   
    df.describe(include='all')
        will give statistical values for all columns(numerical + categorical)   
    df.value_counts()
        will give no of unique rows in the df
    df['col_name].value_counts()
        will give frequency of unique values in the column after removing null values 
    df['col_name].value_counts(normalize= True)
        will give percentage of unique values in the column after removing null values 
    df['col_name].value_counts(normalize= True,dropna=False)
        will give percentage of unique values in the column containing null values 
    df['col_name'].unique()
        will give array of unique values of the column 
    df.loc[start_row_index:end_row_index:row_step,[list of column]]
    df.iloc[start_row_index:end_row_index:row_step,start_col_index:end_col_index:col_step]

    df.dropna() -> to drop na/null values
    fill na/null values
        df.fillna()
        df.bfill()
        df.ffill()
    df.sort_values(by=[list of column])
    df.sort_values(by=[list of column],ascending=[True, False])
    df.sort_index()
    inplace=True => apply the chnages to main dataframe  
    df.drop -> will help drop specific row/column
    df.drop_duplicates -> will help drop duplicate values
